### Игра "Жизнь"

В данном файле созданы четыре класса: рыбы, креветки, скалы и океан. Рыбы, креветки и скалы вынесены в отдельные классы, так как с ними проще работать в таком виде, но эти классы не очень насыщенные. Основным классом выступает океан, который обновляется с помощью функции observation(). Подробнее о том, как задать океан, ниже.

```python
import random
from IPython.display import clear_output
from time import sleep

class Fish:
    
    def __init__(self):
        self.name = 'F'
    
    def __str__(self):
        return self.name
        
'''
'''
        
class Shrimp:
    
    def __init__(self):
        self.name = 'S'
    
    def __str__(self):
        return self.name
        
'''
'''
        
class Cliff:
    
    def __init__(self):

        self.name = 'C'
    
    def __str__(self):
        return self.name
        
'''
'''

class Ocean:
    
    def __init__(self, side_x, side_y, cliff, fish, shrimp):
        self.side_x = side_x
        self.side_y = side_y
        self.cliff = cliff
        self.fish = fish
        self.shrimp = shrimp
        self.ocean = [['0' for i in range(self.side_x)] for i in range(self.side_y)]
        
        for j in range(side_y):
            for i in range(side_x):
                if i == 0 or j == 0 or i == side_x-1 or j == side_y-1:
                    self.ocean[j][i] = 'x'
        
    def print_ocean(self):
        for i in range(self.side_y):
            a = ''
            for j in range(self.side_x):
                a+=str(self.ocean[i][j])+' '
            print(a)
            
   
    def spawn_cliffs(self):
        #for i in range(self.cliff):
        counter = 0
        while counter < self.cliff:
            a = random.randint(0,self.side_y-1)
            b = random.randint(0,self.side_x-1)
            if not type(self.ocean[a][b]) == Cliff and self.ocean[a][b] != 'x':
                self.ocean[a][b] = Cliff()
                counter+=1
                
    def spawn_fish(self):
        counter = 0
        while counter < self.fish:
            a = random.randint(0,self.side_y-1)
            b = random.randint(0,self.side_x-1)
            if self.ocean[a][b] == '0':
                self.ocean[a][b] = Fish()
                counter+=1
                
    def spawn_shrimp(self):
        counter = 0
        while counter < self.shrimp:
            a = random.randint(0,self.side_y-1)
            b = random.randint(0,self.side_x-1)
            if self.ocean[a][b] == '0':
                self.ocean[a][b] = Shrimp()
                counter+=1
                
    def observation(self):

        dead_guys = []
        fish_replication = []
        shrimp_replication = []
        
        for j in range(self.side_y):
            for i in range(self.side_x):
                if type(self.ocean[j][i]) == Fish:
                    counter = -1 ## Так как себя мы впоследствии тоже посчитаем
                    for k in [-1,0,1]:
                        for l in [-1,0,1]:
                            if type(self.ocean[j+k][i+l]) == Fish:
                                counter+=1
                    if (counter < 2) or (counter >= 4):
                        dead_guys.append((j,i))
                        
                elif type(self.ocean[j][i]) == Shrimp:
                    counter = -1 ## Так как себя мы впоследствии тоже посчитаем
                    for k in [-1,0,1]:
                        for l in [-1,0,1]:
                            if type(self.ocean[j+k][i+l]) == Shrimp:
                                counter+=1
                    if (counter < 2) or (counter >= 4):
                        dead_guys.append((j,i))
                        
                elif self.ocean[j][i] == '0':
                    fish_counter = 0
                    shrimp_counter = 0
                    for k in [-1,0,1]:
                        for l in [-1,0,1]:
                            if type(self.ocean[j+k][i+l]) == Fish:
                                fish_counter+=1
                            elif type(self.ocean[j+k][i+l]) == Shrimp:
                                shrimp_counter+=1
                    if fish_counter == 3:
                        fish_replication.append((j,i))
                    elif shrimp_counter == 3:
                        shrimp_replication.append((j,i))
                            
        for j in range(self.side_y):
            for i in range(self.side_x):
                if (j,i) in dead_guys:
                    self.ocean[j][i] = '0'
                    
                elif (j,i) in fish_replication:
                    self.ocean[j][i] = Fish()
                
                elif (j,i) in shrimp_replication:
                    self.ocean[j][i] = Shrimp()
        
    def is_alive(self):
        checker = False
        for j in range(self.side_y):
            for i in range(self.side_x):
                if type(self.ocean[j][i]) == Shrimp or type(self.ocean[j][i]) == Fish:
                    checker = True
                    break
        return checker
```

Чтобы задать параметры океана, вбейте последовательно в ячейке ниже следующие желаемые параметры:
    1. Длина океана по горизонтали
    2. Длина океана по вертикали
    3. Количество скал
    4. Количество рыбы
    5. Количество креветок
    
PS <i>Помните, что сумма числа всех подводных объектов должна быть меньше числа клеток поля</i>

```python
'''
Параметры ниже можно менять
'''
x_side = 20 ## Длина по горизонтали
y_side = 10 ## Длина по вертикали
cliff = 5 ## Количество скал
fish = 40 ## Количество рыбы
shrimp = 40 ## Количество креветок
```

<b>Помните:</b> ячейка ниже прекратит свое выполнение только если океан умрет, то есть в нем не останется рыб и креветок, что маловероятно, так как подводные жители склонны сбиваться в "кучки" по 4 индивида. При необходимости выключите ячейку самостоятельно (ниже можно увидеть, что выполнение ячейки было прервано с клавиатуры) :)

```python
a = Ocean(x_side+2, y_side+2, cliff, fish, shrimp)
a.spawn_cliffs() ## Выставили на поле скалы
a.spawn_fish() ## Выставили рыб
a.spawn_shrimp() ## Выставили креветок

while a.is_alive():
    
    a.print_ocean()
    sleep(3)
    clear_output()
    a.observation()
```
